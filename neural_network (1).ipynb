**Implementing AND Gate in Neural Network **

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

**Assign Input Values**

```python
input_val=np.array([[0,0],[0,1],[1,1],[1,0]])
input_val.shape
```

**Assign Output Values**

```python
output=np.array([0,0,1,0])
output=output.reshape(4,1)
output.shape
```

**Assign Weights**

```python
weights=np.array([[0.1], [0.2]])
weights
```

**Adding Bias**

```python
bias=0.3
```

**Activation Function**

```python
def sigmoid_func(x):
  return 1/(1+np.exp(-x))
```

**Derivative of sigmoid function**

```python
def der(x):
  return sigmoid_func(x) * (1-sigmoid_func(x))
```

**Updating weights**

```python
for epochs in range(10000):
  input_arr = input_val

  weighted_sum = np.dot(input_arr, weights) + bias
  first_output = sigmoid_func(weighted_sum)

  error = first_output - output
  total_error = np.square(np.subtract(first_output,output)).mean()
  #print(total_error)

  first_der = error
  second_der = der(first_output)
  derivative = first_der * second_der

  t_input = input_val.T
  final_derivative = np.dot(t_input,derivative)

  #update weights
  weights=weights - 0.04 * final_derivative

  #update bias
  for i in derivative:
    bias = bias - 0.01 * i

  print(weights)
  print(bias)
```

```python
pred = np.array([0,1])
result=np.dot(pred, weights) + bias
res=sigmoid_func(result)
print(res)
```
